```python
import numpy as np
from matplotlib import pyplot as plt



x, y = np.meshgrid(np.linspace(0,10, 25),
                   np.linspace(0, 10, 25))

dx = 3*y**2
dy = 2*x**2
plt.quiver(x,y,dx,dy,color='g')


# plt.xlim(0, )
# plt.ylim(10,40)
plt.ylabel('y')
plt.xlabel('v')
plt.title('Pred, Prey')
plt.grid()
plt.show()
```
