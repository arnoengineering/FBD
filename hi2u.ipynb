```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
pred_0 = 10
prey_0 = 20

pred = [pred_0]
prey = [prey_0]
pred_dead = -0.5
pred_birth = 0.8

prey_birth = 0.7
prey_dead = -0.8
```

```python
t = np.arange(10)
```

```python
for ti in range(t.size):
    print(f'time {t[ti]} pred{pred[-1]}, prey {prey[-1]}')
    pred_t = pred[-1]*(pred_dead+pred_birth*prey[-1]+1)
    prey_t = prey[-1]*(prey_dead*pred[-1]+prey_birth+1)
    
    pred.append(pred_t)
    prey.append(prey_t)

```

```python
plt.plot(pred,prey)
```
