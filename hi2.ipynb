```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize


space = np.linspace(0,10,20)

x,y = np.meshgrid(space,space)
# x = space
# y=space
```

```python
u = 3*y**2
v=2*x
# print(u)
print(u.shape)
print('\n\n-----')
mag = np.sqrt(u**2+v**2)
print(mag.shape)
norm_mag = Normalize()
norm_mag.autoscale(mag)
print(norm_mag)
col = cm.inferno

cl = np.arange(0,10,2)
```

```python
plt.quiver(x,y,u,v,mag)
for c in cl:
    line_y = (space**2+c**3)**(1/3)
    plt.plot(space, line_y)
```
