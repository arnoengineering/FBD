```python
import matplotlib.pyplot as plt
import sympy as sy
import numpy as np
from sympy import symbols
import matplotlib
sy.init_printing()
s_x,s_y,s_z = symbols('sigma_x sigma_y sigma_z')
t_xy,t_xz,t_yz = symbols('tau_x tau_y tau_z')
mat_1 = sy.Matrix([[s_x, t_xy, t_xz],
         [t_xy, s_y, t_yz],
         [t_xz,t_yz,s_z]])
mat_1
```

```python
fig, ax = plt.subplots(4,3)
f_i = [5, 10, 50, 100]
for i in range(4):
    print('\n-------------\n')
    t1 = np.linspace(0,10,f_i[i])
    func = np.sin(t1*2)+np.cos(5*t1)
    coef = np.fft.rfft(func)
    #print(coef)
    freq = np.fft.rfftfreq(f_i[i],10)
    print(f'size freq1 ({freq.size}) size coef2({coef.size})')
    func_3 = []
    for t in t1:
        e1 = 2*np.pi*1j*freq
        print('e1')
        print(e1)
        e2 = np.exp(e1*t)
        print('e2')
        print(e2)
        e3= coef*e1
        print('e3')
        print(e3)
        e4 = abs(e1)
        print('e4')
        print(e4)
        e5 = np.sum(e4)
        print('e5')
        print(e5)
        func_3.append(e5)
    func_2 = np.array([np.sum(abs(coef*np.exp(2*np.pi*1j*freq*t)/f_i[i])) for t in t1])
    print('size f2', func_2.shape)
    print(func_2)
    ax[i,0].plot(t1,func)
    ax[i,1].plot(t1,func_2)
    ax[i,2].plot(freq, abs(coef))
    print('\n-')
    # ax[i,2].xlim([0,20])
```
